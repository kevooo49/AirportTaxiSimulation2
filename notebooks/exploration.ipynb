# Airport Runway Simulation Exploration

This notebook is used for exploratory data analysis and visualization of the simulation results from the airport runway simulation project.

```python
import pandas as pd
import matplotlib.pyplot as plt
from src.utils import load_runway_data

# Load runway data
runway_data = load_runway_data('data/runway_logs.csv')

# Display the first few rows of the data
runway_data.head()
```

```python
# Visualize airplane movements
plt.figure(figsize=(10, 6))
plt.plot(runway_data['timestamp'], runway_data['airplane_id'], marker='o', linestyle='-')
plt.title('Airplane Movements on Runway')
plt.xlabel('Time')
plt.ylabel('Airplane ID')
plt.grid()
plt.show()
```
